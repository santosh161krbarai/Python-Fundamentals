```python
# question 2
def find_evens(a,b):
    for i in range(a,b):
        if (i+1)%2==0:
            print(i+1)
find_evens(1,10)
```

```python
# question 3
def print_digits(n):
    arr=[]
    while n>0:
        rem=n%10
        arr.append(rem)
        n=n//10
    print("The digits in the given number are ")
    for i in range(len(arr)-1,-1,-1):
        print(arr[i])
print_digits(123)
```

```python
# question 4 and 5
def digit_count(num):
    count=0
    sum=0
    while num>0:
        count+=1
        rem=num%10
        sum+=rem
        num=num//10
    return count,sum
digit_count(123)
```

```python
# question 8
def calculator(a,b,operation):
    match operation:
        case "+":
            return a+b
        case "-":
            return a-b
        case "*":
            return a*b
        case "/":
            return a/b
calculator(10,2,"/")
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
is_prime(4)
```
